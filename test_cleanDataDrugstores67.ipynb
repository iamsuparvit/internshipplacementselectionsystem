```python
# เรียกไฟล์ csv
df = pd.read_csv(file_path, skiprows=3)
```

```python
# เอาแถวล่างที่ว่าง (NaN) ออก
df.dropna(axis=0, how="all", inplace=True)
```

```python
# คัด columns
df = df[['Unnamed: 1','Unnamed: 2','Unnamed: 3','เพศหญิง','เพศชาย', 'ไม่ระบุเพศ','สรุปจำนวนที่รับฝึกได้ในผลัดนี้รวม (คน)', 'เพศหญิง.1', 'เพศชาย.1',
       'ไม่ระบุเพศ.1','สรุปจำนวนที่รับฝึกได้ในผลัดนี้รวม (คน).1']]
df
```

```python
df.rename(columns={
    'Unnamed: 1':'code',
    'Unnamed: 2':'branch',
    'Unnamed: 3':'address',
    'เพศหญิง':'female1',
    'เพศชาย':'male1',
    'ไม่ระบุเพศ':'both1',
    'สรุปจำนวนที่รับฝึกได้ในผลัดนี้รวม (คน)':'seat1',
    'เพศหญิง.1':'female2',
    'เพศชาย.1':'male2',
    'ไม่ระบุเพศ.1':'both2',
    'สรุปจำนวนที่รับฝึกได้ในผลัดนี้รวม (คน).1':'seat2'
}, inplace=True)
df
```

```python
sex_map = {'female1': 'female', 'male1': 'male', 'both1': 'both',
           'female2': 'female', 'male2': 'male', 'both2': 'both'}

df['sex_require1'] = None
for i, row in df.iterrows():
    for col in ['female1', 'male1', 'both1']:
        if row[col] == '/':
            df.at[i, 'sex_require1'] = sex_map[col]
            break

df['sex_require2'] = None
for i, row in df.iterrows():
    for col in ['female2', 'male2', 'both2']:
        if row[col] == '/':
            df.at[i, 'sex_require2'] = sex_map[col]
            break
df[['sex_require1', 'sex_require2']]
```

```python
# แก้โรงพยาบาลำรวจที่ sex_require เป็น None
for i, row in df.iterrows():
    if str(row['code']).startswith('HPT'):
        df.at[i, 'sex_require1'] = 'both'
        df.at[i, 'sex_require2'] = 'both'
df.iloc[18:22]
```

```python
# แก้ sex_require1, sex_require2 เป็น 'both' ถ้า code ขึ้นต้นด้วย 'FCN'
for i, row in df.iterrows():
    if str(row['code']).startswith('FCN'):
        df.at[i, 'sex_require1'] = 'both'
        df.at[i, 'sex_require2'] = 'both'
df
```
